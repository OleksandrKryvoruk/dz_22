Вирішіть задачу: Протягом 30 днів вересня було 12 дощових, 8 вітряних, 4 холодних, 5 дощових та вітряних,
3 дощових та холодних, 2 вітряних та холодних, а один день був дощовий, вітряний та холодний одночасно.
Скільки днів у вересні була гарна погода? Роботу виконайте у середовищі Jupyter Notebook.

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3, venn3_circles

bad_weather = 0
days = 30
weather = {'rain' : 12,
           'wind' : 8,
           'cold' : 4}

intersection = { 'r_w_c' : 1,
                 'r_w' : 5,
                 'r_c' : 3,
                 'w_c' : 2 }

result_weather = {'rain' : [],
                  'wind' : [],
                  'cold' : []}

for i in range(intersection['r_w_c']):
    for weather_day in result_weather:
        result_weather[weather_day].append(bad_weather)
    bad_weather += 1
    for j in weather:
        weather[j] -= 1
    for w in intersection:
        intersection[w] -= 1

for i in range(intersection['r_w']):
    for j in result_weather:
        if j == 'rain' or j == 'wind':
            result_weather[j].append(bad_weather)
            weather[j] -= 1
    bad_weather += 1

for i in range(intersection['r_c']):
    for j in result_weather:
        if j == 'rain' or j == 'cold':
            result_weather[j].append(bad_weather)
            weather[j] -= 1
    bad_weather += 1

for i in range(intersection['w_c']):
    for j in result_weather:
        if j == 'wind' or j == 'cold':
            result_weather[j].append(bad_weather)
            weather[j] -= 1
    bad_weather += 1

for i in weather:
    for j in range(weather[i]):
        result_weather[i].append(bad_weather)
        bad_weather += 1

print(f"September has {days - bad_weather} days with good wather")

venn3([set(result_weather['rain']), set(result_weather['wind']), set(result_weather['cold'])], ('rain', 'wind', 'cold'))
plt.show()
```
